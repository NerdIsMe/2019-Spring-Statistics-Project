```python
df2= pd.read_excel('D:\python statistic\data\DSAP108 videos_sep.xlsx','student id')

print('Head of dataset:')
display(df2.head())

print("Tail of dataset:")
display(df2.tail())
```

```python
from dateutil.rrule import rrule, DAILY,TU
import datetime as dt

results = rrule(DAILY,
        dtstart = dt.datetime(2020,3,1),
        until = dt.datetime(2020, 7,6),
        byweekday = (TU)
)


df4= pd.DataFrame(list(results))
df4
```

```python
df4= pd.DataFrame(results, columns=['date'])
df4
```

```python
df3 = df2['student_id']
df3 = pd.DataFrame(df3)
df3
```

```python
df3['2020-03-03'] = 0
df3['2020-03-10'] = 0
df3['2020-03-17'] = 0
df3['2020-03-24'] = 0
df3['2020-03-31'] = 0
df3['2020-04-07'] = 0
df3['2020-04-14'] = 0
df3['2020-04-21'] = 0
df3['2020-04-28'] = 0
df3['2020-05-05'] = 0
df3['2020-05-12'] = 0
df3['2020-05-19'] = 0
df3['2020-05-26'] = 0
df3['2020-06-02'] = 0
df3['2020-06-09'] = 0
df3['2020-06-16'] = 0
df3['2020-06-23'] = 0
df3['2020-06-30'] = 0
df3
```

```python
df5 = df3.values
df5
```

```python
times_0 = 113623
times_1 = 112
times_2 = 18

for i in  tqdm(range (times_2)):
    for x in  tqdm(range(times_1)):
        
        for y in  tqdm(range(times_0)):
            if i == 18 :
                if  df1['created_at'][y]  >= pd.Timestamp(df4['date'][i], tz=None).to_pydatetime() and  df3['student_id'][x] == df1['student_id'][y] :
                     if df1['end'][y] +1 == df1['start'][y+1]  or df1['end'][y]  == df1['start'][y+1] :
                        df5[x][i+1] = df5[x][i+1] +1
            else :
                
                if  df1['created_at'][y] >= pd.Timestamp(df4['date'][i], tz=None).to_pydatetime() and df1['created_at'][y]  < pd.Timestamp(df4['date'][i+1] , tz=None).to_pydatetime() and df1['student_id'][y] == df3['student_id'][x] :
                                if  df1['end'][y] +1 == df1['start'][y+1]   or df1['end'][y]  == df1['start'][y+1] :
                                       df5[x][i+1] = df5[x][i+1] +1
```

```python
df6 = pd.DataFrame(df5)
df6
```

```python
cols_to_sum =df6.columns[1:df6.shape[1]]
cols_to_sum
```

```python
df6['Total'] = df6[cols_to_sum].sum(axis=1)
df6
```
