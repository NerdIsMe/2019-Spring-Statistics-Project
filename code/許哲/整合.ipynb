```python
import pandas as pd
```

```python
DSAP107 = pd.read_excel('../../data/許哲/DSAP107.xlsx')
DSAP108 = pd.read_excel('../../data/許哲/DSAP108.xlsx')
OR107 = pd.read_excel('../../data/許哲/OR107.xlsx')
```

```python
DSAP107_p = pd.read_excel('../../data/data2.0/pause_frequency/pause_frequency_week_DSAP 107.xlsx')
DSAP108_p = pd.read_excel('../../data/data2.0/pause_frequency/pause_frequency_week_DSAP 108.xlsx')
OR107_p = pd.read_excel('../../data/data2.0/pause_frequency/pause_frequency_week_OR 107.xlsx')
```

```python
DSAP107_p = DSAP107_p.rename(columns = {0:'student_id', 'Total':'pause_freq'})
DSAP108_p = DSAP108_p.rename(columns = {0:'student_id', 'Total':'pause_freq'})
OR107_p = OR107_p.rename(columns = {0:'student_id', 'Total':'pause_freq'})
```

```python
DSAP107 = DSAP107.merge(DSAP107_p[['student_id', 'pause_freq']])
DSAP108 = DSAP108.merge(DSAP108_p[['student_id', 'pause_freq']])
OR107 = OR107.merge(OR107_p[['student_id', 'pause_freq']])
```

```python
# save data
DSAP107.to_excel('../../data/許哲/DSAP107.xlsx', index = False)
DSAP108.to_excel('../../data/許哲/DSAP108.xlsx', index = False)
OR107.to_excel('../../data/許哲/OR107.xlsx', index = False)
```
